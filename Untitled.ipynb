```python
df = df.T
df.head()
```

```python
def find_recipe(ingredients):
    df["Score"] = 0
    for ingr in ingredients:
        for recipe in df["ingredients"]:
            for recipe_ingredient in recipe:
                if ingr in recipe_ingredient:
                    return recipe
```
